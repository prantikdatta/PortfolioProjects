```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('Downloads/Movie Industry/movies.csv')
```

```python
df.head()
```

```python
#checking for missing values in dataframe

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#data types?

df.dtypes
```

```python
df1 = df.fillna(0)
```

```python
#changing data type to int to make it more easily readable

df1['budget'] = df1['budget'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
df1['votes'] = df1['votes'].astype('int64')
```

```python
df1.head()
```

```python
df1 = df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#drop duplicates

df1.drop_duplicates().head()
```

```python
#Scatter plot budget vs gross

plt.scatter(x=df1['budget'], y=df1['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df1.head()
```

```python
#budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df1, scatter_kws={'color':'red'}, line_kws={'color':'black'})
```

```python
#correlation in data

#default = pearson
df1.corr(method='pearson') #kendall, spearman
```

```python
correlation_matrix = df1.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for numeric features")
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#looking at the features with names and trying to assign them some values
```

```python
df_numeric = df1
for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes  #assigning random values based on the feature identity
        
df_numeric.head()
```

```python
df1.head()
```

```python
correlation_matrix = df_numeric.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for numeric features")
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numeric.corr()
```

```python
corr_mat = df_numeric.corr()
corr_pairs = corr_mat.unstack()   #unstacking -> used to understand correlation within larger feature matrix
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
